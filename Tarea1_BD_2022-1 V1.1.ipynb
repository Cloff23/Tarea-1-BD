## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Usuarios"

```python
%%sql
create table Usuarios(
    ID_usuario integer GENERATED ALWAYS AS IDENTITY,
    nombre varchar(45) NOT NULL,
    apellido varchar(45),
    correo varchar(45) NOT NULL,
    contraseña varchar(50) NOT NULL,
    fecha_registro timestamp NOT NULL,
    PRIMARY KEY(ID_usuario)
);
```

### Tabla "Creadores"

```python
%%sql
create table Creadores(
    ID_creador integer GENERATED ALWAYS AS IDENTITY,
    nombre varchar(45) NOT NULL,
    apellido varchar(45) NOT NULL,
    popularidad integer,
    Estilo varchar(45),
    PRIMARY KEY(ID_creador)
);
```

### Tabla "Tipos_NTF"

```python
%%sql
create table tipos_NFT(
    ID_tipo integer GENERATED ALWAYS AS IDENTITY,
    nombre varchar(20) NOT NULL,
    PRIMARY KEY(ID_tipo)
);
```

### Tabla "NFT"

```python
%%sql
create table NFT(
    ID_nft integer GENERATED ALWAYS AS IDENTITY,
    nombre varchar(45) NOT NULL,
    ID_creador integer NOT NULL,
    ID_tipo integer,
    PRIMARY KEY(ID_nft),
    FOREIGN KEY(ID_creador) REFERENCES Creadores(ID_creador),
    FOREIGN KEY(ID_tipo) REFERENCES tipos_NFT(ID_tipo)
);
```

### Tabla "Precio NFT"

```python
%%sql
Create table Precio_NFT(
    ID_nft integer,
    fecha timestamp NOT NULL,
    precio integer NOT NULL,
    PRIMARY KEY(ID_nft, fecha),
    FOREIGN KEY(ID_nft) REFERENCES NFT(ID_nft)
);
```

### Tabla "Usuario tiene NFT"

```python
%%sql
Create table Usuario_tiene_NFT(
    id_usuario integer,
    id_nft integer,
    PRIMARY KEY(id_usuario, id_nft),
    FOREIGN KEY(id_usuario) REFERENCES Usuarios(ID_usuario),
    FOREIGN KEY(id_nft) REFERENCES NFT(ID_nft)
);
```

### Consulta 2
**Cantidad de precios que ha tenido históricamente cada NTF.**

Ejemplo:


| NFT     | Cantidad de precios |
|--------------|----------|
| Minecraft Sphere |    3     |
| The Binary Girl     |    7     |

**Solución:** se unen las tablas "precio_nft" y "nft", mediante las id's de nft, para seleccionar el nombre del nft y contar sus precios.

```python
%%sql
SELECT
	nft.nombre AS "NFT", COUNT(precio_nft.id_nft) AS "Cantidad de precios"
FROM
	nft
INNER JOIN precio_nft
ON nft.id_nft = precio_nft.id_nft
GROUP BY nft.nombre;
```

**El nombre y tipo (nombre del tipo) del NFT históricamente más caro.**
###### Si hay más de uno, ponga solamente el primero

**Solución:** Se unen las tablas "nft" y "tipos_nft", mediante el id del tipo, donde el id del nft esta dado por el id que tenga maximo precio en la tabla "precio_nft". Ademas, se limita a 1 fila.

```python
%%sql
SELECT
	nft.nombre AS "NFT", tipos_nft.nombre AS "Tipo"
FROM
	nft
INNER JOIN tipos_nft ON nft.id_tipo = tipos_nft.id_tipo
WHERE
	nft.id_nft = (SELECT
				  	id_nft
				  FROM
				  	precio_nft
				  WHERE
				  	precio = (SELECT
							 	MAX(precio)
							 FROM
							 	precio_nft
							 )
				 )
LIMIT 1;
```

### Consulta 7
**Top 4 NFTs más caros actualmente de un id_tipo X .**

Ejemplo:

|NFT | 
|--------|
|Minecraft Sphere | 
|The Binary Girl |
|Koalas in Love | 
|Art Dodger | 

**Solución:** Se selecciona nombre de "nft" de la tabla resultante dada por la seleccion del id de "precio_nft", de la fecha máxima de "precio_nft", es decir, fecha mas reciente registrada y se le llamará "precio_nft_actuales"
Luego se unira el id del precio de "precio_nft" con el id del precio de "precios_nft_actuales" y la fecha de "precios_nft" con la fecha de "precio_nft".
Despues se unira el id del nft de "nft" con el id del nft de "precios_nft_actuales"
Posteriormente se unirá el id del tipo nft de "tipos_nft" con el id del tipo de "nft"
Finalmente se agruparan mostrando el nombre del nft, la fecha, el precio y el nombre del nft "X", el cual será introducido manualmente , se mostraran los 4 mayores precios mediante un orden de precios descendientes y un limite de 4.

```python
%%sql
SELECT
	nft.nombre AS "NFT"
FROM
	(SELECT
		id_nft, MAX(fecha) AS fecha
	FROM
		precio_nft
	GROUP BY ID_NFT) AS precios_nft_actuales
INNER JOIN precio_nft ON precio_nft.id_nft = precios_nft_actuales.id_nft AND precios_nft_actuales.fecha = precio_nft.fecha
INNER JOIN nft ON nft.id_nft = precios_nft_actuales.id_nft
INNER JOIN tipos_nft ON tipos_nft.id_tipo = nft.id_tipo
GROUP BY nft.nombre, precio_nft.fecha, precio_nft.precio, tipos_nft.nombre
HAVING tipos_nft.nombre = 'Imagen'
ORDER BY precio DESC
LIMIT 4;
```

### Consulta 9 
**Creador al cual le han comprado más NFTs.**

Ejemplo:
   
|Nombre |Apellido |Cantidad |
|-|--|-----|
| Levi |Ackerman |32|

**Solución:** Se unen las tablas "creadores" y "nft", mediante el id de creador, luego a esta misma tabla se une la tabla "usuario_tiene_nft", mediante el id del nft en ésta tabla y en la "nft". Se seleccionan el nombre y apellido de la tabla "creadores" y se cuenta la cantidad de usuarios que poseen un nft de éste creador. Se ordena la tabla resultante de mayor a menor, por medio de la cantidad descrita anteriormente. Finalmente, se limita a una fila.

```python
%%sql
SELECT
	creadores.nombre AS "Nombre", creadores.apellido AS "Apellido", COUNT(usuario_tiene_nft.id_nft) AS "Cantidad"
FROM 
	creadores
INNER JOIN nft ON nft.id_creador = creadores.id_creador
INNER JOIN usuario_tiene_nft ON usuario_tiene_nft.id_nft = nft.id_nft
GROUP BY creadores.nombre, creadores.apellido
ORDER BY COUNT(usuario_tiene_nft.id_nft) DESC
LIMIT 1;
```

### Consulta 10  
**El NFT con mayor precio actual, que pertenezca a un creador X.**

Ejemplo:
   
|NFT |Valor|
|-|-------|
|Minecraft Sphere |18000|

**Solución:** Se selecciona el nombre y el precio de la tabla dada por la seleccion del id del nft de "precio_nft" y con la fecha maxima, es decir, la fecha mas actual de "precio_nft" y se le llamara a la tabla resultante "precios_nft_actuales". 
Despues se unira el id del nft de "precio_nft" con el id del nft de "precio_nft_actuales" y la fecha de "precio_nft_actuales" con la fecha de "precio_nft". 
Luego se unira el id del nft de "nft" con el id del nft de "precios_nft_actuales"
Posteriormente se uniran el id del creador de "creadores" con el id de del creador de "nft"
Finalmente se agruparan mostrando el nombre del nft, la fecha el precio y el nombre del creador "X", el cual será introducido manualmente , se mostrará el que tiene precio mayor mediante un orden de precios descendientes y un limite de 1.

```python
%%sql
SELECT
	nft.nombre AS "NFT", precio_nft.precio AS "Valor"
FROM
	(SELECT
		id_nft, MAX(fecha) AS fecha
	FROM
		precio_nft
	GROUP BY ID_NFT) AS precios_nft_actuales
INNER JOIN precio_nft ON precio_nft.id_nft = precios_nft_actuales.id_nft AND precios_nft_actuales.fecha = precio_nft.fecha
INNER JOIN nft ON nft.id_nft = precios_nft_actuales.id_nft
INNER JOIN creadores ON creadores.id_creador = nft.id_creador
GROUP BY nft.nombre, precio_nft.fecha, precio_nft.precio, creadores.nombre
HAVING creadores.nombre = 'Marvin'
ORDER BY precio DESC
LIMIT 1;
```
